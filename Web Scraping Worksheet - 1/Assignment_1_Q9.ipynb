Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title,
company name, CTC, and apply date.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get("https://internshala.com/fresher-jobs")
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
job_title = soup.find_all("div",class_="heading_4_5 profile")
job_title
```

```python
title_list = []
for i in job_title:
        title_list.append(i.text.replace("\n",""))
title_list
```

```python
companies = soup.find_all("a",class_="link_display_like_text")
companies
```

```python
company_list = []
for i in companies:
    company_list.append(i.text.strip())
    
company_list
```

```python
other_details = soup.find_all("div",class_="internship_other_details_container")
other_details
```

```python
other_details_list = []
for i in other_details:
    temp = i.text.split("\n")
    temp_list = []
    for j in temp:
        j = j.strip()
        if(j != ""):
            temp_list.append(j)
    other_details_list.append(temp_list)
other_details_list
```

```python
ctc_list = []
apply_date_list = []
for i in other_details_list:
    ctc_list.append(i[3])
    apply_date_list.append(i[5])
print(ctc_list)
apply_date_list
```

```python
df = pd.DataFrame({})
df["Job Title"] = title_list
df["Company"] = company_list
df["CTC"] = ctc_list
df["Apply By"] = apply_date_list
df
```
