```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
soup.prettify()
```

```python
headers = soup.find_all(["h1","h2","h3","h4","h5","h6"])
headers
```

```python
header_list = []
for i in headers:
        header_list.append(i.text.strip().replace("\n",""))

header_list
```
