# 6. Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:

i) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.

ii) Top 10 women’s ODI players along with the records of their team and rating.

iii) Top 10 women’s ODI all-rounder along with the records of their team and rating.

```python
from bs4 import BeautifulSoup
import requests
```

```python
import pandas as pd
import numpy as np
```

# i) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.

```python
page = requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi")
page
```

```python
soup = BeautifulSoup(page.content)
print(soup.prettify())
```

```python
team_details = soup.find_all("tr",class_=["rankings-block__banner","table-body"])
team_details
```

```python
team_detail_list = []
for j in team_details:
    temp = j.text.split("\n")
    temp_list = []
    for k in temp:
        k = k.strip()
        if(k != ""):
            temp_list.append(k)
    team_detail_list.append(temp_list)    
team_detail_list = np.array(team_detail_list)
team_detail_list
```

```python
DF = pd.DataFrame(team_detail_list[:,[1,3,4,5]],columns=["Team","Matches","Points","Rating"])
DF
```

# ii) Top 10 women’s ODI players along with the records of their team and rating.

```python
page2 = requests.get("https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting")
page2
```

```python
soup2 = BeautifulSoup(page2.content)
print(soup2.prettify())
```

```python
player_details = soup2.find_all("tr",class_=["rankings-block__banner","table-body"],limit=10)
player_details
```

```python
player_name_list = []
for i in player_details:
    for j in i.find_all("td",class_="table-body__cell rankings-table__name name") or i.find_all("div",class_="rankings-block__banner--name-large"):
        player_name_list.append(j.text.replace("\n",""))
player_name_list
```

```python
player_country_list = []
for i in player_details:
    for j in i.find_all("span",class_="table-body__logo-text") or i.find_all("div",class_="rankings-block__banner--nationality"):
        player_country_list.append(j.text.replace("\n","").strip())
player_country_list
```

```python
player_rating_list = []
for i in player_details:
    for j in i.find_all("td",class_="table-body__cell rating") or i.find_all("div",class_="rankings-block__banner--rating"):
        player_rating_list.append(j.text.replace("\n","").strip())
player_rating_list
```

```python
team_record_list = []
for i in player_country_list:
    for j in range(len(team_detail_list)):
        if(i==team_detail_list[j][2]):
            indices = [1,3,4,5]
            team_record_list.append(team_detail_list[j][indices])
            break
team_record_list = np.array(team_record_list)
team_record_list
```

```python
DF2 = pd.DataFrame({"Player Name":player_name_list,
                    "Country Name":team_record_list[:,0],
                    "Matches Played":team_record_list[:,1],
                    "Points":team_record_list[:,2],
                    "Team Rating":team_record_list[:,3],
                    "Player Rating":player_rating_list})
DF2
```
