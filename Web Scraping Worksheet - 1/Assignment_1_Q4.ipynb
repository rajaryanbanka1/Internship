```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://bookpage.com/reviews")
page.content
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
soup.prettify
```

```python
book_article = soup.find_all("div",class_="flex-article-content")
book_article = book_article[:5]
book_article
```

```python
book_names = []
for i in book_article:
    for j in i.find_all("h4",class_="italic"):
        for k in j.find_all("a",href=True):
            temp = k.text.replace("★","")
            book_names.append(temp.strip())
book_names
```

```python
#author name, genre and book review
author_names = soup.find_all("p",class_="sans bold")
author_names = author_names[:5]
author_names
```

```python
author_name_list = []
for i in author_names:
    author_name_list.append(i.text.replace("\n",""))
author_name_list
```

```python
genres = soup.find_all("p",class_="genre-links hidden-phone")
genres = genres[:5]
genres
```

```python
genre_list = []
for i in genres:
    genre_list.append(i.text.replace("\n",""))
genre_list
```

```python
page_list = []
for i in book_article:
        for j in i.find_all("h4",class_="italic"):
            for k in j.find_all("a",href=True):
                page_list.append("https://bookpage.com"+k["href"])
page_list = page_list[:5]
page_list
```

```python
book_reviews=[]

for i in page_list:
    page_temp = requests.get(i)
    soup_temp = BeautifulSoup(page_temp.content)
    book = soup_temp.find_all("div",class_="article-body")
    for j in book:
        book_reviews.append(j.text.replace("\n",""))
    

book_reviews
```

```python
import pandas as pd

data = pd.DataFrame()
```

```python
data["Book Name"] = book_names
data["Author Name"] = author_names_list
data["Genre"] = genre_list
data["Book Review"] = book_reviews

data.pd.set_option()
```
