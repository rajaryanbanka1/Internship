Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:

i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

ii) Top 10 ODI Batsmen in men along with the records of their team and rating.

iii) Top 10 ODI bowlers along with the records of their team and rating.

```python
from bs4 import BeautifulSoup
import requests
```

# i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
page = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
import pandas as pd
```

```python
team_details = soup.find_all("tr",class_=["table-body","rankings-block__banner"])
team_details
```

```python
team_detail_list = []
for i in team_details:
    temp = i.text.split("\n")
    temp_list = []
    for j in temp:
        j = j.strip()
        if(j != ""):
            temp_list.append(j.strip())
    team_detail_list.append(temp_list)
team_detail_list
```

```python
names = []
matches = []
points = []
ratings = []
for i in range(10):
    names.append(team_detail_list[i][1])
    matches.append(team_detail_list[i][3])
    points.append(team_detail_list[i][4])
    ratings.append(team_detail_list[i][5])
```

```python
data_df = pd.DataFrame()
data_df["Team Name"] = names
data_df["Matches"] = matches
data_df["Points"] = points
data_df["Ratings"] = ratings
data_df
```

# ii) Top 10 ODI Batsmen in men along with the records of their team and rating.

```python
page2 = requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting")
page2
```

```python
soup2 = BeautifulSoup(page2.content)
soup2
```

```python
player_details = soup2.find_all("tr",class_=["rankings-block__banner","table-body"],limit=10)
player_details
```

```python
player_detail_list = []
for i in player_details:
    temp = i.text.split("\n")
    temp_list = []
    for j in temp:
        j = j.strip()
        if(j != ""):
            temp_list.append(j.strip())
    player_detail_list.append(temp_list)
player_detail_list
```

```python
import numpy as np
player_detail_list1 = np.array(player_detail_list)
print(player_detail_list1)
player_detail_list1[:,2]
```

```python
team_detail_list1 = np.array(team_detail_list)
print(team_detail_list1)
team_detail_list1[:,2]
```

```python
player_team_records = []
count = 0
for i in range(len(player_detail_list1)):
    for j in range(len(team_detail_list1)):
        if(player_detail_list1[i][2]==team_detail_list1[j][2]):
            player_team_records.append(list(player_detail_list1[i,1:3]))
            for k in team_detail_list1[j,3:6]:
                player_team_records[count].append(k)
            player_team_records[count].append(player_detail_list1[i,3])
            count+=1
#            Team_records.append(np.concatenate(player_detail_list1[i],team_detail_list1[j]))
#           Team_records.append(temp_list)
       
player_team_records
```

```python
data = pd.DataFrame(player_team_records,columns=["Name","Team","Matches","Points","Team Rating","Player Rating"])
data
```
