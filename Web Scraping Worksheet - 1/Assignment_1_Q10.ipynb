# Write a python program to scrape house details from https://www.nobroker.in/ for any location. It should include house title, location, area, emi and price

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get("https://www.nobroker.in/property/sale/bangalore/Kormangla%20Bangalore?searchParam=W3sibGF0IjoxMi45NDE2OTM1LCJsb24iOjc3LjYxNjgxOTYsInBsYWNlSWQiOiJDaElKMTdxXzVrZ1VyanNSMTFsMUZHTkE1Y0UiLCJwbGFjZU5hbWUiOiJLb3JtYW5nbGEgQmFuZ2Fsb3JlIn1d&radius=2.0")
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
house_titles = soup.find_all("h2",class_="heading-6 font-semi-bold nb__1AShY")
house_titles
```

```python
house_title_list = []
for i in house_titles:
    house_title_list.append(i.text)
house_title_list
```

```python
locations = soup.find_all("div",class_="nb__2CMjv")
locations
```

```python
location_list = []
for i in locations:
    location_list.append(i.text)
location_list
```

```python
other_details = soup.find_all("div",class_="nb__17R6o")
other_details
```

```python
detail_list = []
for i in other_details:
    temp_list = []
    for j in i.find_all("div",class_="font-semi-bold heading-6"):
        temp_list.append(j.text)
    detail_list.append(temp_list)
detail_list
```

```python
#area, emi and price
```

```python
import numpy as np
```

```python
detail_list = np.array(detail_list)
detail_list
```

```python
df = pd.DataFrame({})
df["House Title"] = house_title_list
df["Location"] = location_list
df["Area"] = detail_list[:,0]
df["EMI"] = detail_list[:,1]
df["Price"] = detail_list[:,2]
df
```
