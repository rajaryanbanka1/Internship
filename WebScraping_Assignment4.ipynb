```python
for i in url_list:
    driver.get(i)
    
    dataset_name_tag=driver.find_element_by_xpath("//span[@class='heading']")
    dataset_name_list.append(dataset_name_tag.text)
    
    task_tag = driver.find_element_by_xpath("//table[@cellpadding='6']/tbody/tr[3]/td[2]/p")
    task_list.append(task_tag.text)
    
    data_type_tag = driver.find_element_by_xpath("//td/p[4]")
    data_type_list.append(data_type_tag.text)
    
    attribute_type_tag = driver.find_element_by_xpath("//td/p[6]")
    attribute_type_list.append(attribute_type_tag.text.replace("\n",". "))
    
    no_of_instance_tag = driver.find_element_by_xpath("//table[@cellpadding='6']/tbody/tr/td[4]/p")
    no_of_instance_list.append(no_of_instance_tag.text)
    
    no_of_attribute_tag = driver.find_element_by_xpath("//table[@cellpadding='6']/tbody/tr[2]/td[4]/p")
    no_of_attribute_list.append(no_of_attribute_tag.text)
    
    year_tag = driver.find_element_by_xpath("//table[@cellpadding='6']/tbody/tr[2]/td[6]/p")
    year_list.append(year_tag.text[:year_tag.text.find("-")])
```

```python
df = pd.DataFrame({})
df["Dataset Name"] = dataset_name_list
df["Data Type"] = data_type_list
df["Task"] = task_list
df["Attribute Type"] = attribute_type_list
df["No. of Instances"] = no_of_instance_list
df["No. of Attribute"] = no_of_attribute_list
df["Year"] = year_list
df
```
